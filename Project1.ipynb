```python
# Identify how many duplicates there are
df.duplicated().sum()
```

```python
# Look at the duplicate values
duplicates = df[df.duplicated(keep=False)]
duplicates
```

```python
# Remove values that are duplicates across all columns.

df = df.drop_duplicates(keep="last")

df.duplicated().sum() # Make sure this is equal to zero now.
```

```python
to_keep = ["HadHeartAttack", "BMI", "SleepHours", "PhysicalActivities", "AlcoholDrinkers", "GeneralHealth", "ECigaretteUsage", "SmokerStatus"]
```

```python
# Remove irrelevant columns; those which are not lifestyle choices/general health. 

df = df[to_keep]
```

```python
print(df.isna().sum()) # Which columns have na values and how many
```

```python
nancolumns = to_keep

df_subset = df[nancolumns]
```

```python
# Remove values that are null across all columns of interest (and therefore contribute no information).
# Find how many values will be removed.
print(df_subset.isnull().all(axis=1).sum())
df = df.dropna(subset=nancolumns, how="all")
```

Investigating unique values for various categorical columns to ensure consistency.

```python
categorical_cols = ["HadHeartAttack", "PhysicalActivities", "AlcoholDrinkers", "GeneralHealth", "ECigaretteUsage", "SmokerStatus"]

for c in categorical_cols:
    print(c, ":", df[c].unique())
```

```python
# Changing the order of GeneralHealth to a scale from worst to best.

order = ["Poor", "Fair", "Good", "Very good", "Excellent"]

df["GeneralHealth"] = pd.Categorical(df["GeneralHealth"], categories=order, ordered=True)
```

Changing SmokerStatus and ECigaretteUsage to shorter category names that will match for later comparison.

```python
df["SmokerStatus"] = df["SmokerStatus"].replace("Never smoked", "Never")
df["SmokerStatus"] = df["SmokerStatus"].replace("Current smoker - now smokes some days", "Some days")
df["SmokerStatus"] = df["SmokerStatus"].replace("Former smoker", "Former")
df["SmokerStatus"] = df["SmokerStatus"].replace('Current smoker - now smokes every day', "Every day")
```

```python
# Changing the order of SmokerStatus to a scale from lest to most.

order = ["Never", "Former", "Some days", "Every day"]

df["SmokerStatus"] = pd.Categorical(df["SmokerStatus"], categories=order, ordered=True)
```

```python
df["ECigaretteUsage"] = df["ECigaretteUsage"].replace("Not at all (right now)", "Former")
df["ECigaretteUsage"] = df["ECigaretteUsage"].replace("Never used e-cigarettes in my entire life", "Never")
df["ECigaretteUsage"] = df["ECigaretteUsage"].replace("Use them some days", "Some days")
df["ECigaretteUsage"] = df["ECigaretteUsage"].replace("Use them every day", "Every day")
```

```python
# Changing the order of ECigaretteUsage to a scale from lest to most.
# Same order as for SmokerStatus
df["ECigaretteUsage"] = pd.Categorical(df["ECigaretteUsage"], categories=order, ordered=True)
```

```python
custom_palette2 = ["#E9328B", "#fce9eb", "#920B4C","#C6186C", "#ED4D9A", "#F163A8", "#F67BB6", "#FF8DC4", "#99D6B3"]
custom_palette3 = sns.color_palette("blend:#fdcbd5,#fa4b6e", 25)
```

```python
fig, ax = plt.subplots(figsize=(20,10))

sns.boxplot(data=df, y="HadHeartAttack", x="BMI", color="#E9328B", ax=ax)
plt.title("Distribution of BMI for HadHeartAttack")
plt.show()
```
